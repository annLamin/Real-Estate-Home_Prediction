```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head(5)
```

```python
# Removed columns that are not relevant to our prediction
```

```python
df.drop(['date','condition', 'view', 'waterfront','statezip', 'country','sqft_above','sqft_basement','street'], axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df1 = df
```

```python
df1.info()
```

```python
# These columns cannot be float but integers
columns_to_convert = ['bedrooms', 'bathrooms','floors']
df1[columns_to_convert] = df1[columns_to_convert].apply(lambda x: np.ceil(x).astype(int))

df1.head(5)
```

```python
df1.isnull().sum()
```

```python
# Checking the cities as they are categorical. we will apply dummy variables and reduce the dimensions
```

```python
df1.city.unique()
```

```python
city_stats = df1['city'].value_counts(ascending=False)
city_stats
```

```python
len(city_stats[city_stats<=5])
```

```python
city_stats_below_5 = city_stats[city_stats<=5]
city_stats_below_5
```

```python
df1.city = df1.city.apply(lambda x: 'other' if x in city_stats_below_5 else x)
len(df1.city.unique())
```

```python
df1.city.value_counts()
```

```python
df1.head(5)
```

```python
df2 = df1
```

```python
# Outlier Removal Using Bathrooms Feature
```

```python
df2
```

```python
df2[df2.bathrooms>3]
```

```python
# It is obviously unusual to have more than 1 bathroom than the number of beedrooms so we will take that as an outlier
```

```python
df2[df2.bathrooms>=df2.bedrooms+2]
```

```python
df3 = df2[df2.bathrooms<df2.bedrooms+2]
df3.shape
```

```python
# we cannot also have the compound bigger than the living room therefore we see tham as outliers
```

```python
df3[df3.sqft_living>df3.sqft_lot]
```

```python
df4 = df3[df3.sqft_living<df3.sqft_lot]
```

```python
df4.shape
```

```python
def remove_price_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('city'):
        m = np.mean(subdf.price)
        st = np.std(subdf.price)
        reduced_df = subdf[(subdf.price>(m-st)) & (subdf.price<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df5 = remove_price_outliers(df4)
df5.shape
```

```python
# creating dummy variables for the city column.
```

```python
dummies = pd.get_dummies(df5.city)
dummies.head(5)
```

```python
df6 = pd.concat([df5, dummies.drop('other', axis='columns')], axis='columns')
df6.head(5)
```

```python
df7 = df6.drop('city', axis='columns')
df7.head(5)
```

```python
# Building Our Model
```

```python
X = df7.drop('price', axis='columns')
X.head(5)
```

```python
X.shape
```

```python
y = df7.price
y.head(5),
```

```python
 y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,  y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test, y_test)
```

```python

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
# class sklearn.linear_model.LinearRegression(*, fit_intercept=True, copy_X=True, n_jobs=None, positive=False)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': { 
                'copy_X' : [True, False],
                'fit_intercept' : [True, False],
                'n_jobs' : [1,2,3],
                'positive' : [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(city,bedrooms,bathrooms,sqft_living,sqft_lot,floors,yr_built,yr_renovated):    
    loc_index = np.where(X.columns==city)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = bedrooms
    x[1] = bathrooms
    x[2] = sqft_living
    x[3] = sqft_lot
    x[4] = floors
    x[5] = yr_built
    x[6] = yr_renovated
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```
